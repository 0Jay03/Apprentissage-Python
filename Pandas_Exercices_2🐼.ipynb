### Découpage d'une chaîne de caractères avec `split`

Une date comme `'28-02-2014'` est une chaîne où le jour, le mois et l'année sont séparés par un tiret (`-`). La classe des chaînes de caractères en Python propose la méthode `split`, qui permet de découper une chaîne selon un séparateur spécifique :

```python
date = '28-02-2014'

# Découpage de la chaîne sur le caractère '-'
print(date.split('-'))
>>> ['28', '02', '2014']


````python
date = '28-02-2014'
# Découpage de la chaîne sur le caractère '-'
print(date.split('-')) 
````

````python
import pandas as pd

link = 'https://raw.githubusercontent.com/QuantikDataStudio/dataset/main/Transactions.csv'

transactions = pd.read_csv(link, sep =',', index_col = "transaction_id")

# Suppression des doublons
transactions = transactions.drop_duplicates(keep = 'first')

# Supprimer les lignes avec des "/" dans la colonne "tran_date"
transactions = transactions[~transactions['tran_date'].str.contains('/')]

transactions
````

+ Définir une fonction get_day prenant en argument une chaîne de caractères et qui renvoie le premier élément de son découpage par le caractère '-'.

+ Définir les fonctions get_month et get_year qui font de même avec le deuxième et troisième élément du découpage.

+ Dans 3 variables days, months et years, stocker le résultat de la méthode apply sur la colonne tran_date appliquée avec les fonctions get_day, get_month et get_year. Comme ces fonctions s'appliquent élément par élément, il n'est pas nécessaire de spécifier l'argument axis dans la méthode apply.

+ Créer les colonnes 'day', 'month' et 'year' dans le DataFrame et y stocker les valeurs de days, months et years. La création d'une nouvelle colonne se fait simplement en la déclarant :


```
# Création d'une nouvelle colonne 'day' avec les valeurs contenue dans days.

```

+ Afficher les 5 premières lignes de transactions.

````python
def get_day(date):           # Définition des fonctions à appliquer à la colonne 'tran_date' pour les jours 
    splits = date.split('-')
    day = splits[0]
    return day
````

````python
def get_month(date):         # Définition des fonctions à appliquer à la colonne 'tran_date' pour les mois
    splits = date.split('-')
    month = splits[1]
    return month
````

````python
def get_year(date):              # Définition des fonctions à appliquer à la colonne 'tran_date' pour les annees
    splits = date.split('-')
    year = splits[2]
    return year
````

````python
days = transactions['tran_date'].apply(get_day)
````

````python
months = transactions['tran_date'].apply(get_month)
````

````python
years = transactions['tran_date'].apply(get_year)
````

````python
# Créer les colonnes 'day', 'month' et 'year' dans le DataFrame et y stocker les valeurs de days, months et years.
transactions['day'] = days
transactions['month'] = months
transactions['year'] = years

transactions.head(5)
````

## Gestion des valeurs manquantes

Dans un des exercices précédents, nous avons utilisé la méthode replace de transactions pour remplacer les valeurs manquantes par 0.
Cette approche manque de rigueur et il ne faut pas procéder de cette façon en pratique.

Pour cette raison, nous allons réimporter la version brute de transactions pour annuler les étapes que nous avons faites dans les exercices précédents.



---



**Dans la réalite, il est rare, voir impossible, d'avoir un DataFrame ou une source de données propre et sans NaNs.
Mais c'est le cas dans ce DataFrame que nous utilisons depuis le début de la session.
De fait, nous allons volontairement et aléatoirement supprimer 20% des valeurs afin de réaliser les prochains exercices.**

````python
import pandas as pd
import numpy as np

link = 'https://raw.githubusercontent.com/QuantikDataStudio/dataset/main/Transactions.csv'

transactions = pd.read_csv(link, sep=',', index_col="transaction_id")
transactions = transactions.drop_duplicates(keep='first')
new_names = {'Store_type': 'store_type',
             'Qty': 'qty',
             'Rate': 'rate',
             'Tax': 'tax'}

transactions = transactions.rename(new_names, axis=1)
transactions = transactions[~transactions['tran_date'].str.contains('/')]

# Colonnes spécifiées où les valeurs seront remplacées
selected_columns = ['prod_cat_code', 'qty', 'rate', 'tax', 'total_amt', 'store_type']

# Calcul du nombre total de valeurs à remplacer
total_values = transactions[selected_columns].size

# Calcul du nombre de valeurs à remplacer par NaN (environ 10%)
num_values_to_remove = int(0.2 * total_values)

# Génération aléatoire des indices de valeurs à supprimer
random_indices = np.random.randint(0, total_values, num_values_to_remove)

# Remplacement des valeurs aux indices sélectionnés par NaN
for idx in random_indices:
    col_idx = idx % len(selected_columns)
    row_idx = idx // len(selected_columns)
    col_name = selected_columns[col_idx]
    transactions.loc[transactions.index[row_idx], col_name] = np.nan

transactions
````

### Code generique 

````python
# Cette fonction permet de déterminer les variables avec des valeurs manquantes
def valeur_manquante(df):
    flag=0
    for col in df.columns:
            if df[col].isna().sum() > 0:
                flag=1
                print(f'"{col}": {df[col].isna().sum()} valeur(s) manquante(s)')
    if flag==0:
        print("Le dataset ne contient plus de valeurs manquantes.")

valeur_manquante(transactions)
````
