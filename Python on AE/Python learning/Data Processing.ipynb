```python
#EXERCICES SUR LA FONCTION CONCAT

import pandas as pd
import numpy as np

link = 'https://raw.githubusercontent.com/QuantikDataStudio/dataset/main/Transactions.csv'

transactions = pd.read_csv(link, sep=',', index_col="transaction_id")
transactions = transactions.drop_duplicates(keep='first')
new_names = {'Store_type': 'store_type',
             'Qty': 'qty',
             'Rate': 'rate',
             'Tax': 'tax'}

transactions = transactions.rename(new_names, axis=1)
transactions.info()
```

```python
# Séparer les variables du DataFrame transactions en deux avec la moitié des variables dans un DataFrame nommé part_1 et la deuxième moitié dans un DataFrame nommé part_2.

# Nombre total de colonnes
total_columns = len(transactions.columns)

# Trouver le point de séparation
mid_point = total_columns // 2  # Divise en deux parties égales

# Création des deux DataFrames
part_1 = transactions.iloc[:, :mid_point]  # Première moitié des colonnes
part_2 = transactions.iloc[:, mid_point:]  # Deuxième moitié des colonnes
```

```python
# Reconstituer transactions dans un DataFrame nommé union en concaténant part_1 et part_2.

union = pd.concat([part_1, part_2], axis=1)

union
```

```python
# Que se passe-t-il si on concatène part_1 et part_2 en renseignant l'argument axis = 0 ?

union = pd.concat([part_1, part_2], axis=0)

union
```

```python
# EXERCICES Fusion de DataFrames avec la méthode merge

# Le DataFrame customer contient des informations sur des clients de la colonne 'cust_id' de transactions.

# La colonne 'customer_Id' du DataFrame customer permettra de faire la jointure entre transactions et customer. Cela permettra d'enrichir le jeu de données transactions avec des informations supplémentaires

# À l'aide de la méthode rename et d'un dictionnaire, renommer la colonne 'customer_Id' du DataFrame customer par 'cust_id'.

customer = customer.rename(columns={'customer_Id': 'cust_id'})
```

```python
# En utilisant la méthode merge, effectuer la jointure à gauche entre les DataFrames transactions et customer sur la variable 'cust_id'. Nommer le DataFrame créé fusion.

fusion = transactions.merge(right = customer, on = 'cust_id', how='left')

print(fusion.columns)
```

```python
#CONCAT VS MERGE
# PART 1: CONCAT (utilise soit la ligne soit la colonne pour concatener des dataframes)
import pandas as pd

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

# Concaténation par ligne (par défaut axis=0)
result = pd.concat([df1, df2], axis=0)
print(result)
```

```python
#CONCAT VS MERGE
# PART 2: MERGE (utilise une cle et le type de jointure pour concatener des dataframes)

import pandas as pd

df1 = pd.DataFrame({'ID': [1, 2], 'A': [3, 4]})
df2 = pd.DataFrame({'ID': [1, 3], 'B': [5, 6]})

# Fusion des DataFrames sur la colonne "ID"
result = pd.merge(df1, df2, on='ID', how='inner')
print(result)
```

```python
# Trier et ordonner les valeurs d'un df

# Lancer cette celulle

# Definition des dictionnaires
data_bateaux = {'nom_bateau'  : ['Julia', 'Siren', 'Sea Sons', 'Hercules', 'Cesar', 'Minerva'],
                'couleur'    : ['bleu', 'vert', 'rouge', 'bleu', 'jaune', 'vert'],
                'numero_reservation': [2, 3, 6, 1, 4, 5],
                'nombre_reservations': [34, 10, 20, 41, 12, 16]}

data_clients = {'id_client' : [91, 154, 124, 320, 87, 22],
                'nom_client'        : ['Marie', 'Anna', 'Yann', 'Lea', 'Marc', 'Yassine'],
                'id_reservation': [1, 2, 3, 7, 9, 5]}

# Creation des DataFrames
bateaux = pd.DataFrame(data_bateaux)
clients = pd.DataFrame(data_clients)
```

Nous voulons déterminer facilement quel client a réservé les bateaux du DataFrame bateaux.

Pour cela, il suffit de fusionner les DataFrames.

```python
# Nous voulons déterminer facilement quel client a réservé les bateaux du DataFrame bateaux.
# 
# Pour cela, il suffit de fusionner les DataFrames.

# Renommer la colonne 'numero_reservation' de bateaux en 'id_reservation' grâce à la méthode rename.
bateaux_renamed = bateaux.rename(columns={'numero_reservation': 'id_reservation'})
bateaux_renamed
```

```python
# Dans un DataFrame nommé bateaux_clients, faire la jointure à gauche entre bateaux et clients.

bateaux_clients = pd.merge(bateaux_renamed, clients, on='id_reservation', how="left")

# Définir la colonne 'nom_bateau' comme index du DataFrame bateaux_clients
bateaux_clients.set_index('nom_bateau', inplace=True)
bateaux_clients
```

```python
#À l'aide de la méthode loc qui permet d'indexer un DataFrame, trouver qui a réservé les bateaux 'Julia' et 'Siren'.

# JE trouve les donnees relies a ces bateaux
resultats = bateaux_clients.loc[['Julia', 'Siren']]

# Je trouve le nom des clients relies a cs bateaux
noms_resultats = resultats['nom_client']
print(noms_resultats)
```

```python
# À l'aide de la méthode isna appliquée sur la colonne nom_client, déterminer les bateaux qui n'ont pas été réservés.
bateaux_non_reserves = bateaux_clients[bateaux_clients['nom_client'].isna()]
print(bateaux_non_reserves)
```

```python
# Le nombre de fois qu'un bateau a été réservé jusqu'à présent est renseigné par la colonne 'nombre_reservations'. À l'aide de la méthode sort_values, déterminer le nom du client qui a réservé le bateau bleu ayant le plus de réservations à son actif


# Filtrer les bateaux de couleur 'bleu'
bateaux_bleus = bateaux_clients[bateaux_clients['couleur'] == 'bleu']

# Trier les bateaux bleus par 'nombre_reservations' de manière décroissante
bateaux_bleus_tries = bateaux_bleus.sort_values(by='nombre_reservations', ascending=False)

# Trouver le nom du client ayant réservé le bateau bleu avec le plus de réservations
client_avec_plus_reservations = bateaux_bleus_tries.iloc[0]['nom_client']

# Afficher le résultat
print(client_avec_plus_reservations)
```

```python
# Grouper les éléments d'un DataFrame: méthodes groupby, agg et crosstab.

#À l'aide d'une opération groupby, déterminer pour chaque client à partir de la quantité d'items achetés dans une transaction (colonne qty) :

import numpy as np

link = 'https://raw.githubusercontent.com/QuantikDataStudio/dataset/main/Transactions.csv'

transactions = pd.read_csv(link, sep=',', index_col="transaction_id")

transactions = transactions.drop_duplicates(keep='first')
new_names = {'Store_type': 'store_type',
             'Qty': 'qty',
             'Rate': 'rate',
             'Tax': 'tax'}

transactions = transactions.rename(new_names, axis=1)


n_qty = lambda qty: len(np.unique(qty))

functions_to_apply = {
    'total_amt' : ['min', 'max', 'sum'],
    'store_type' : n_qty
}

transactions.groupby('cust_id').agg(functions_to_apply)
```

```python
# La quantité maximale.
# La quantité minimale.
# La quantité médiane.


import pandas as pd

# Exemple de données
transactions = pd.DataFrame({
    'cust_id': [1, 1, 2, 2, 3, 3],
    'qty': [5, 8, 0, 2, 7, 1]
})

# Quantité maximale
max_qty = lambda qty: qty[qty > 0].max()

# Quantité minimale
min_qty = lambda qty: qty[qty > 0].min()

# Quantité médiane
median_qty = lambda qty: qty[qty > 0].median()

# Définition du dictionnaire de fonctions à appliquer
functions_to_apply = {
    'qty': [max_qty, min_qty, median_qty]
}

# Opération groupby
qty_groupby = transactions.groupby('cust_id').agg(functions_to_apply)

# Renommer les colonnes produites par le groupby
qty_groupby.columns.set_names(['metric', 'stat'], inplace=True)

# Afficher les premiers résultats
print(qty_groupby.head())
```
