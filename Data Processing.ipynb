```python
#EXERCICES SUR LES OPERATEURS (FILTRATION DES DONNEES)

import pandas as pd
import numpy as np

link = 'https://raw.githubusercontent.com/QuantikDataStudio/dataset/main/Transactions.csv'

transactions = pd.read_csv(link, sep=',', index_col="transaction_id")
transactions = transactions.drop_duplicates(keep='first')
new_names = {'Store_type': 'store_type',
             'Qty': 'qty',
             'Rate': 'rate',
             'Tax': 'tax'}

transactions = transactions.rename(new_names, axis=1)
transactions.info()
```

```python
# À partir de transactions, créer un DataFrame nommé e_shop contenant uniquement les transactions effectuées dans les magasins de type 'e-Shop' avec un montant total supérieur à 5000 (colonnes 'store_type' et 'total_amt').

e_shop = transactions[(transactions['store_type'] == 'e-Shop') & (transactions['total_amt'] > 5000)]

#POur compter le nombre d'element dans la colonne
e_shop.shape[0]
```

```python
# De même, créer une DataFrame nommé teleshop qui contient les transactions effectuées dans les magasins de type 'TeleShop' avec un montant total de plus de 5000.
teleshop = transactions[(transactions['store_type'] == 'Teleshop') & (transactions['total_amt'] > 5000)]

#POur compter le nombre d'element dans la colonne
teleshop.shape[0]
```

```python
import pandas as pd

# URLs des fichiers CSV
prod_url = 'https://raw.githubusercontent.com/QuantikDataStudio/dataset/main/prod_cat_info.csv'
customer_url = 'https://raw.githubusercontent.com/QuantikDataStudio/dataset/main/Customer.csv'

# Importer les fichiers dans des DataFrames
prod_cat_info = pd.read_csv(prod_url)
customer = pd.read_csv(customer_url)

# Afficher un aperçu des données
print(prod_cat_info.head())  # Affiche les premières lignes du DataFrame prod_cat_info
print(customer.head())       # Affiche les premières lignes du DataFrame customer
```

```python
# Les colonnes Gender et city_code de customer contiennent deux valeurs manquantes chacune. Les remplacer par leur mode à l'aide des méthodes fillna et mode.
# Remplacer les valeurs manquantes par le mode de chaque colonne
for col in ['Gender', 'city_code']:
    mode_value = customer[col].mode()[0]  # Trouve la valeur la plus fréquente
    customer[col] = customer[col].fillna(mode_value)  # Remplace les NaN par le mode
```

```python
#EXERCICES SUR LA FONCTION CONCAT

import pandas as pd
import numpy as np

link = 'https://raw.githubusercontent.com/QuantikDataStudio/dataset/main/Transactions.csv'

transactions = pd.read_csv(link, sep=',', index_col="transaction_id")
transactions = transactions.drop_duplicates(keep='first')
new_names = {'Store_type': 'store_type',
             'Qty': 'qty',
             'Rate': 'rate',
             'Tax': 'tax'}

transactions = transactions.rename(new_names, axis=1)
transactions.info()
```

```python
# Séparer les variables du DataFrame transactions en deux avec la moitié des variables dans un DataFrame nommé part_1 et la deuxième moitié dans un DataFrame nommé part_2.

# Nombre total de colonnes
total_columns = len(transactions.columns)

# Trouver le point de séparation
mid_point = total_columns // 2  # Divise en deux parties égales

# Création des deux DataFrames
part_1 = transactions.iloc[:, :mid_point]  # Première moitié des colonnes
part_2 = transactions.iloc[:, mid_point:]  # Deuxième moitié des colonnes
```

```python
# Reconstituer transactions dans un DataFrame nommé union en concaténant part_1 et part_2.

union = pd.concat([part_1, part_2], axis=1)

union
```

```python
# Que se passe-t-il si on concatène part_1 et part_2 en renseignant l'argument axis = 0 ?

union = pd.concat([part_1, part_2], axis=0)

union
```

```python
# EXERCICES Fusion de DataFrames avec la méthode merge

# Le DataFrame customer contient des informations sur des clients de la colonne 'cust_id' de transactions.

# La colonne 'customer_Id' du DataFrame customer permettra de faire la jointure entre transactions et customer. Cela permettra d'enrichir le jeu de données transactions avec des informations supplémentaires

# À l'aide de la méthode rename et d'un dictionnaire, renommer la colonne 'customer_Id' du DataFrame customer par 'cust_id'.

customer = customer.rename(columns={'customer_Id': 'cust_id'})
```

```python
# En utilisant la méthode merge, effectuer la jointure à gauche entre les DataFrames transactions et customer sur la variable 'cust_id'. Nommer le DataFrame créé fusion.

fusion = transactions.merge(right = customer, on = 'cust_id', how='left')

print(fusion.columns)
```

```python
#Est-ce que la fusion a produit des NaNs ?
print(fusion[['cust_id']].isnull().sum())  # Doit afficher 0 pour chaque colonne
```
